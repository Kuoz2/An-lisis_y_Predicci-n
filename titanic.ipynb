```python
pasajeros_id_148= df[df["PassengerId"]==148]
pasajeros_id_148
```

```python
porcen_supervivos= df['Survived'].value_counts(normalize=True)*100
porcen_supervivos
```

```python
supervi_clases= df.groupby('Pclass')['Survived'].mean()*100
supervi_clases
```

```python
edad_mediana_mujeres= df[df['Sex'] == 'female'].groupby('Pclass')['Age'].mean()
edad_mediana_mujeres
```

```python
df_edad = df[df['Age'].notnull]
es_max = df_edad['Age'] >= 18
sobrevivencia = df_edad.groupby(['Pclass', es_max])['Survived'].mean() * 100
sobrevivencia
```
