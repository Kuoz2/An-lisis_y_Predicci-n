````python
def tipo_mas_fuerte(df):

    df_temp = df.copy()

    df_temp['Poder_TCG'] = df_temp['HP'] + df_temp[['Attack', 'Sp. Atk']].max(axis=1)



    tipo_ganador = df_temp.groupby('Type 1')['Poder_TCG'].mean().idxmax()

    return tipo_ganador

    #pass
````

````python
busqueda_de_pokemon = input("Ingresa el nombre del pokemon")

def buscar_pokemon(df, nombre):

    resultado = df[df['Name'].str.contains(nombre, case=False, na=False)]

    return resultado

    #pass

tipo1 = input("Ingresa un tipo de pokemon").lower()

tipo2 = input("Ingresa un tipo de pokemon").lower()

if tipo2 == tipo1:

  print("Este tipo de pokemon ya fue ingresado. Por favor, ingresa otro tipo")

  tipo2 = input("Ingresa un tipo de pokemon")
````

````python
def comparar_tipos(df, tipo1, tipo2):

    t1 = tipo1.capitalize()

    t2 = tipo2.capitalize()



    stats = ['HP', 'Attack', 'Speed']



    res1 = df[df['Type 1'] == t1][stats].mean().round(2)

    res2 = df[df['Type 1'] == t2][stats].mean().round(2)



    dict_comparativo = {

        t1: res1.to_dict(),

        t2: res2.to_dict(),

        'Ganadores': {

            'HP': t1 if res1['HP'] > res2['HP'] else t2,

            'Attack': t1 if res1['Attack'] > res2['Attack'] else t2,

            'Speed': t1 if res1['Speed'] > res2['Speed'] else t2

        }

    }

    return dict_comparativo

    #pass
````

````python
print("\n TTPO MÁS FUERTE")

print("-" * 20)

print(tipo_mas_fuerte(df))

print("-" * 20)

print("\n BUSCAR POKEMON")

print("-" * 20)

print(buscar_pokemon(df, busqueda_de_pokemon))

print("-" * 20)

resultado = comparar_tipos(df, tipo1, tipo2)

print(f"\n RESUMEN DE DUELO: {tipo1.upper()} vs {tipo2.upper()}")

print("-" * 45)

for stat, ganador in resultado['Ganadores'].items():

    val1 = resultado[tipo1.capitalize()][stat]

    val2 = resultado[tipo2.capitalize()][stat]



    print(f"En {stat:6}: Gana {ganador:8} ⮕ ({val1} vs {val2})")

print("-" * 45)
````
