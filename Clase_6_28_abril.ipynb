# 🔹 Ejercicio L1 — Estadísticas sin funciones built-in

Dada la lista: `temperaturas = [23.1, 19.8, 25.4, 17.2, 30.0, 22.5, 18.9, 27.3, 21.1, 24.6]`

**Sin usar** `max()`, `min()`, `sum()`:
1. Calcula el promedio
2. Encuentra el valor máximo y mínimo
3. Cuenta cuántos valores están sobre el promedio
4. Imprime un reporte con los resultados

````python
temperaturas = [23.1, 19.8, 25.4, 17.2, 30.0, 22.5, 18.9, 27.3, 21.1, 24.6]

def calcular_promedio(lista):
  suma = 0
  for numero in lista:
    suma = suma + numero
  return round(suma / len(lista),2)

print(f"Promedio: {calcular_promedio(temperaturas)}")
````

````python
def encontrar_maximo(lista):
  maximo = lista[0]
  for numero in lista:
    if numero > maximo:
      maximo = numero
  return maximo

print(f"Maximo: {encontrar_maximo(temperaturas)}")
````

````python
def encontrar_minimo(lista):
  minimo = lista[0]
  for numero in lista:
    if numero < minimo:
      minimo = numero
  return minimo

print(f"Minimo: {encontrar_minimo(temperaturas)}")
````
