Usando el dataset de Pokémon, responde:

1. ¿Cuántos pokémon son legendarios?

2. ¿Cuántos pokémon tienen un `Total` mayor a 500?

3. ¿Cuántos pokémon de tipo Psíquico (`Psychic`) hay en la Generación 1?

4. ¿Cuál es el pokémon más rápido (`Speed`) que NO sea legendario?

5. Lista los 5 tipos con más representantes (`Type 1`)

```python
import pandas as pd
```

```python
df = pd.read_csv("/content/drive/MyDrive/CURSOPYTHON/pokemon.csv")
```

```python
df.columns
```

```python
legen = df['Legendary'].sum()
legen
```

```python
total_500 = df[df['Total'] > 500].shape[0]
total_500
```

```python
poke_psico= df[(df['Type 1'] == 'Psychic')  & (df['Generation'] == 1)].shape[0]
poke_psico
```

```python
poke_rapid_no_legen = df[df['Legendary'] == False].sort_values("Speed", ascending=False).head(1)
poke_rapid_no_legen
```

```python
top_5_tipos = df['Type 1'].value_counts().head(5)
top_5_tipos
```
