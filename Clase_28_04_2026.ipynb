## Ejercicio L1 — Estadísticas sin funciones built-in

Dada la lista: `temperaturas = [23.1, 19.8, 25.4, 17.2, 30.0, 22.5, 18.9, 27.3, 21.1, 24.6]`

**Sin usar** `max()`, `min()`, `sum()`:
1. Calcula el promedio
2. Encuentra el valor máximo y mínimo
3. Cuenta cuántos valores están sobre el promedio
4. Imprime un reporte con los resultados


### 1. Promedio

```python
temp = [23.1, 19.8, 25.4, 17.2, 30.0, 22.5, 18.9, 27.3, 21.1, 24.6]

# 1. Promedio
def promedio(valores):

    return sum(valores)/len(valores)

promedio(temp)
```

### 2. Maximo y minimo

```python
# 2. Maximo y minimo
def maximo(valores):
    max_val = valores[0]
    for x in valores:
      if x > max_val:
        max_val = x
    return max_val

def minimo(valores):
    min_val = valores[0]
    for x in valores:
      if x < min_val:
        min_val = x
    return min_val

print(f"Maxima Temp:{maximo(temp)}")
print(f"Minima Temp:{minimo(temp)}")
```

### 3. Valores sobre promedio

```python
prom_temp = promedio(temp)
def mayor_prom(valores):
  mayor_que = []
  for x in valores:
    if x > prom_temp:
      mayor_que.append(x)
  return mayor_que

mayor_prom(temp)
```

###4. Reporte

```python
max_temp = maximo(temp)
min_temp = minimo(temp)
valores_sobre_promedio = mayor_prom(temp)

print("--- Reporte de Temperaturas ---")
print(f"1. Promedio de temperaturas: {prom_temp:.2f}")
print(f"2. Temperatura máxima: {max_temp:.2f}")
print(f"   Temperatura mínima: {min_temp:.2f}")
print(f"3. Valores sobre el promedio ({prom_temp:.2f}): {valores_sobre_promedio}")
print("-------------------------------")
```

## Ejercicio C1 — Calculadora de impuestos

En Chile, el impuesto a la renta se calcula por tramos (2024):

| Tramo | Renta mensual | Tasa |
|-------|--------------|------|
| 1 | Hasta $772.000 | 0% |
| 2 | $772.001 - $1.715.000 | 4% |
| 3 | $1.715.001 - $2.858.000 | 8% |
| 4 | $2.858.001 - $3.975.000 | 13.5% |
| 5 | Más de $3.975.000 | 23% |

Escribe un programa que:
1. Pida la renta mensual con `input()`
2. Calcule el impuesto y la renta líquida
3. Muestre un resumen claro con el tramo, tasa, impuesto y líquido

```python
# 1. Pide la renta mensual con input()
renta_mensual = int(input("Ingrese su renta mensual en CLP: "))
```

```python
# 2. Calcule el impuesto y la renta líquida

impuesto = 0
tasa = "0%"
tramo = "1"

if renta_mensual <= 772000:
    impuesto = 0
    tasa = "0%"
    tramo = "1"
elif renta_mensual <= 1715000:
    impuesto = (renta_mensual - 772000) * 0.04
    tasa = "4%"
    tramo = "2"
elif renta_mensual <= 2858000:
    impuesto = (renta_mensual - 1715000) * 0.08 + (1715000 - 772000) * 0.04 # This calculation needs to be precise with accumulated taxes
    tasa = "8%"
    tramo = "3"
elif renta_mensual <= 3975000:
    impuesto = (renta_mensual - 2858000) * 0.135 + (2858000 - 1715000) * 0.08 + (1715000 - 772000) * 0.04
    tasa = "13.5%"
    tramo = "4"
else:
    impuesto = (renta_mensual - 3975000) * 0.23 + (3975000 - 2858000) * 0.135 + (2858000 - 1715000) * 0.08 + (1715000 - 772000) * 0.04
    tasa = "23%"
    tramo = "5"

renta_liquida = renta_mensual - impuesto
renta_liquida
```

```python
# 3. Muestra un resumen claro con el tramo, tasa, impuesto y líquido

print("\n--- Resumen Calculadora de Impuestos ---")
print(f"Renta Mensual Bruta: ${renta_mensual:,.0f}")
print(f"Tramo de Impuesto: {tramo}")
print(f"Tasa Aplicada: {tasa}")
print(f"Impuesto Calculado: ${impuesto:,.0f}")
print(f"Renta Líquida: ${renta_liquida:,.0f}")
print("----------------------------------------")
```
